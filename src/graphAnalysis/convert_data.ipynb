```python
import json
```

```python
#读取原始数据放入字典中，进行清洗，只留下爬到了关注信息的up,并且顺便去除没有关注任何人的up
with open("up_meta.json") as fp:
    meta_dict=json.load(fp)
```

```python
final_dict={}
num_of_up=0
for key,value in meta_dict.items():
    if 'follow' in value.keys() and 'type' in value.keys():
        if len(value['follow']):
            num_of_up+=1
            final_dict[key]=value
        
print(num_of_up) 
len(final_dict)
```

```python
print(final_dict)
```

```python
#这一步把所有关注信息写入，格式：第一行为总人数，之后每行，第一个是该up的id,第二个是该up的名字，第三个是up的类别（发视频最多的分区),之后每个是关注数，为了方便读取，每一行以0结尾。由于uid的大小不超过整型，用整型储存，顺便把type化成数字存进字典里

type_dict={}#分区字典
with open("up_info.txt","w",encoding='utf-8') as fp:
    fp.write(str(len(final_dict))+'\n')
    for key,value in final_dict.items():
        line=''
        line+=key+' '
        line+=value['name']+' '
        if value['type'] not in type_dict.keys():
            type_dict[value['type']]=len(type_dict)
        line+=str(type_dict[value['type']])+' '
        for follow in value['follow']:
            line+=str(follow)+' '
        line+='0\n'
        fp.write(line)
print(type_dict)

```
